```python
import bs4
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib notebook
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import re
```

```python
url = 'http://thechive.com/2016/04/21/what-you-need-to-make-to-live-comfortably-in-the-50-biggest-us-cities-50-photos/?utm_source=facebook&utm_medium=post&utm_term=lifestyle&utm_campaign=livecomfortably&_branch_match_id=204643810230209045'
```

```python
webtoread=urlopen(url)
soup=BeautifulSoup(webtoread,"lxml")
```

```python
text_source=soup.find('div',{'id': 'gallery-1'})
```

```python
location_vector=[]
Total_vector=[]
Necessities_vector=[]
Spending_vector=[]
Saving_vector=[]
data=[]
for states_num, item in enumerate(text_source.contents[0]):
    text= ''.join(item.findAll(text=True)).split()
    data = filter(None,text)
    data = list(data)
    Location_index = list(data).index('Income')
    if states_num != 1:
        locations = ' '.join(data[1:Location_index])
    else:
        locations = ' '.join(data[Location_index-2:Location_index])
    locations = locations.replace(':','')
    location_vector.append(locations)
    Total_index = data.index('needed:')
    Total_needed = data[Total_index+1]
    Total_needed = re.sub('[\$,]','',Total_needed)
    Total_vector.append(Total_needed)

    Necessities_index = data.index('necessities:')
    Necessities = data[Necessities_index+1]
    Necessities = re.sub('[\$,]','',Necessities)
    Necessities_vector.append(Necessities)
    
    Spending_index = data.index('spending:')
    Spending = data[Spending_index+1]
    Spending = re.sub('[\$,]','',Spending)
    Spending_vector.append(Spending)
    
    Saving_index = data.index('savings:')
    Saving = data[Saving_index+1]
    Saving = re.sub('[\$,]','',Saving)
    Saving_vector.append(Saving)

datatoframe = {'Cities':location_vector,'Total needed':Total_vector,
               'Necessities':Necessities_vector,'Spendings':Spending_vector,
               'Savings':Saving_vector}
```

```python
data = pd.DataFrame(datatoframe)
```

```python
for col_name in ['Total needed','Necessities','Spendings','Savings']:
    data[col_name]=data[col_name].apply(float)
```

```python
data = data.sort(['Total needed'])
```

```python
bb=data[['Necessities','Spendings','Savings']].plot.barh(stacked=True)
bb.set_yticklabels(data['Cities'])
plt.xlabel('Costs per year ($)')
plt.ylabel('50 Cities')
plt.title('Costs in living these cities, 50% for necessities, 30% for spendings, and 20% for savings')
```
